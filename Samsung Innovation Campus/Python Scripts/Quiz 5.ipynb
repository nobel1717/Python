```python
m2 = 0
def insertionsort(S,m2):
    n = len(S)
    for i in range(1, n):
        print(S)
        x = S[i]
        j = i - 1
        print("no")
        while j >= 0 and S[j] > x:
            S[j + 1] = S[j]
            j -= 1
            m2 = m2 + 1
            print("Si")
            print(m2)
        S[j +1] = x
            

S = [50,30,40,10,20]

insertionsort(S,m2)
print(S)
```
